### Importar dependencias

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import math
```

### Leer datos

```python
screen_width = 0
screen_height = 0
target_number = 0
target_x = 0
target_y = 0
x_coordinates = []
y_coordinates = []

lineas_leidas = 0

with open('eye_tracker_data_1.txt') as file:
    for i, item in enumerate(file):
        linea = item.split(" ")
        
        # Get screen dimensions
        if (i == 0):
            screen_width = int(linea[1])
            screen_height = int(linea[3])
        
        # Get target coordinates
        elif (i == 1):
            target_number = linea[1]
            target_x = float(linea[3])
            target_y = float(linea[5])
            
        else:
            new_x = linea[1]
            new_y = linea[3]
            x_coordinates.append(float(new_x))
            y_coordinates.append(float(new_y))
            
            
        lineas_leidas+=1
        
        
            
print(lineas_leidas , "lineas leidas")
print(len(x_coordinates), "puntos obtenidos")
print(target_x, target_y)
```

### Mostrar datos

```python
plt.scatter(x_coordinates, y_coordinates)

plt.xlim(-screen_height, screen_width)
plt.ylim(screen_height, -screen_height)

plt.xlabel("X")
plt.ylabel("Y")
plt.title("Target: " + target_number)
plt.scatter(target_x, target_y, color='red');
plt.show()
```

## Definir funci√≥n que haga toda la chamba

```python
def showDataFrom(fileName, save = False):
    screen_width = 0
    screen_height = 0
    target_number = 0
    target_x = 0
    target_y = 0
    x_coordinates = []
    y_coordinates = []

    suma_distancias = 0

    with open(fileName) as file:
        for i, item in enumerate(file):
            linea = item.split(" ")

            # Get screen dimensions
            if (i == 0):
                screen_width = int(linea[1])
                screen_height = int(linea[3])

            # Get target coordinates
            elif (i == 1):
                target_number = linea[1]
                target_x = float(linea[3])
                target_y = float(linea[5])

            else:
                new_x = float(linea[1])
                new_y = float(linea[3])
                
                if (new_x < 0):
                    new_x = 0
                elif (new_x > screen_width):
                    new_x = screen_width
                
                if (new_y < 0):
                    new_y = 0
                elif (new_y > screen_height):
                    new_y = screen_height
                    
                x_coordinates.append(new_x)
                y_coordinates.append(new_y)
                
                suma_distancias += math.sqrt( math.pow(target_x - new_x,2) + math.pow(target_y - new_y, 2))



    print(len(x_coordinates), "puntos obtenidos")
    print("objetivo: ", target_number)
    print("localizado en ", target_x, ",", target_y)

    plt.scatter(x_coordinates, y_coordinates)

    plt.xlim(0, screen_width)
    plt.ylim(screen_height, 0)

    plt.xlabel("X")
    plt.ylabel("Y")
    plt.title("Target: " + target_number)
    plt.scatter(target_x, target_y, color='red');
    plt.show()
    print("Distancia promedio al objetivo: ", suma_distancias / len(x_coordinates))
```
